# LR预测鸢尾花

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
X, y = load_iris(return_X_y=True)
y
```

```python
# 查看数据缺失率
pd.merge(
    pd.DataFrame(X).isnull().sum().reset_index(),
    (pd.DataFrame(X).isnull().sum()/pd.DataFrame(X).shape[1]).reset_index(),
    how='left',
    on='index'
).rename(columns={'index':'字段名', '0_x':'缺失记录数', '0_y':'缺失率'})
```

```python
# 分离训练集测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.3)
```

```python
# 训练
clf = LogisticRegression()
clf.fit(X_train, y_train)
# 预测
preds = clf.predict(X_test)
# 评估
print(classification_report(y_test, preds, np.unique(y)))
```

```python
# y原值和预测值，打印出来看下
pd.concat([pd.DataFrame(y_test, columns=['y_test']), pd.DataFrame(preds, columns=['preds'])], axis=1)
```
